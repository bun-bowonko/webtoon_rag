```python
import pandas as pd

# 엑셀 파일 읽기
file_path = "../sfx.xlsx"  # 읽을 파일 경로
df = pd.read_excel(file_path, engine="openpyxl")

# 데이터 확인
print(df.head())  # 첫 5개 행 출력
```

```python
print(df['Korean'])
```

```python
print(df['English'])
```

```python
import pandas as pd
nan_cnt = 0
for i in df['English']:
    if pd.isna(i):
        nan_cnt += 1
print(nan_cnt)
```

```python
en = []
ko = []

for i in range(len(df)):
    if not pd.isna(df['Korean'][i]) and not pd.isna(df['English'][i]): 
        ko.append(df['Korean'][i])
        en.append(df['English'][i])
        
print(len(en), len(ko))

data = []

for i, pair in enumerate(zip(ko, en)):
    e_pair = pair[1]
    if ':' in e_pair:
        if '\n' in e_pair:
            lines = e_pair.split('\n')
            filtered_item = []
            for line in lines:
                filtered_item.append(line.split(':')[1])
            data.append(f"{pair[0]} > {','.join(filtered_item)}".strip())
        else:
            data.append(f"{pair[0]} > {pair[1].split(':')[1]}".strip())
    else:
        if '\n' in e_pair:
            lines = e_pair.split('\n')
            data.append(f"{pair[0]} > {','.join(lines)}".strip())
        else:
            data.append(f"{pair[0]} > {pair[1]}".strip())
```

```python
for d in data:
    print(d)
```

```python
with open(f"../sfx_dict.txt", 'w', encoding='utf-8') as file:
    file.writelines('\n'.join(data))
```

```python
def add_text(text):
    lines = text.split("\n")
    en = []
    ko = []
    for i,line in enumerate(lines):
        components = line.split('	')

        if len(components)>0:
            en.append(components[0])
            ko.append(components[1])
    return en, ko
```

```python
def add_text2(text):
    lines = text.split("\n")
    en = []
    ko = []
    for i,line in enumerate(lines):
        components = line.split('	')

        if len(components)>0:
            en.append(components[0])
            ko.append(components[1])
            en.append(components[2])
            ko.append(components[3])
    return en, ko
```

```python
def add_text3(text):
    lines = text.split("\n")
    en = []
    ko = []
    
    for i in range(len(lines)):
        try:
            e = lines[i*6]
            k = lines[i*6+2]
            #print(k,e)
            ko.append(k)
            en.append(e)
        except:
            pass

    return en,ko
```
